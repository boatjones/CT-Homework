```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('data/Taylor_Train.csv', encoding = 'ISO-8859-1')
df.head()
```

```python
len(df.columns)
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_').str.replace('(','').str.replace(')','')
df.head()
```

```python
df.columns = df.columns.str.replace('_/','')
df.head()
```

```python
df[['tickets_sold', 'tickets_available']] = df['attendance_tickets_sold_available'].str.split('/', expand=True)
df.head()
```

```python
df.drop(columns='attendance_tickets_sold_available', inplace=True)
df.head()
```

```python
df['opening_acts'].value_counts()
```

```python
df[['opening_act1', 'opening_act2', 'opening_act3']] = df['opening_acts'].str.split('\r\n', expand=True)
df.head()
```

```python
df.drop(columns='opening_acts', inplace=True)
df.head()
```

```python
df['opening_act1'].isna().sum()
```

```python
df['opening_act1'] = df['opening_act1'].fillna('None')
df['opening_act2'] = df['opening_act2'].fillna('None')
df['opening_act3'] = df['opening_act3'].fillna('None')
```

```python
df['opening_act3'].value_counts()
```

```python
df['opening_act1'] = df['opening_act1'].replace('\x97','None')
```

```python
df.info()
```

```python
# df[df.columns[1:]] = df[df.columns[1:]].replace('[\$,]', '', regex=True).astype(float)
df['revenue'] = df['revenue'].replace('[\$,]', '', regex=True).str.strip()
```

```python
df['revenue'] = df['revenue'].replace('\x97',-1)
```

```python
df['revenue'] = df['revenue'].replace('NaN',-1)
```

```python
df[df['revenue'].isnull()]
```

```python
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')
```

```python
df['revenue'] = df['revenue'].fillna(value=-1)
```

```python
df['revenue'] = df['revenue'].astype(int)
```

```python
df['revenue'].value_counts()
```

```python
df['tickets_sold'] = df['tickets_sold'].fillna(value=-1)
```

```python
df['tickets_sold'] = df['tickets_sold'].replace('[\,]', '', regex=True).str.strip()
```

```python
df['tickets_sold'] = df['tickets_sold'].fillna(value=-1)
```

```python
df['tickets_sold'] = df['tickets_sold'].replace('\x97',-1)
```

```python
df['tickets_sold'] = df['tickets_sold'].astype(int)
```

```python
df['tickets_available'] = df['tickets_available'].replace('[\,]', '', regex=True).str.strip()
df['tickets_avaiable'] = df['tickets_available'].replace('\x97',-1)
df['tickets_available'] = df['tickets_available'].fillna(value=-1)
```

```python
df['tickets_available'] = df['tickets_available'].astype(int)
```

```python
'''
For London at O2 Arena for the same tour, the numbers can be assumed to be the same for the two null rows
The following SQL is necessary to be done in Python:

update df 
set revenue = 5829240, tickets_sold = 74740, tickets_available = 74740, revenue_per_ticket = 77.993578
where venue = 'The O2 Arena' and city = 'London' and tour = 'The_Red_Tour' and revenue = -1 
'''

update_condition = (df['venue'] == 'The O2 Arena') & (df['city'] == 'London') & (df['tour'] == 'The_Red_Tour') & (df['revenue'] == -1)
            
df.loc[update_condition, 'revenue'] = 5829240
df.loc[update_condition, 'tickets_sold'] = 74740
df.loc[update_condition, 'tickets_available'] = 74740
df.loc[update_condition, 'revenue_per_ticket'] = 77.993578
```
